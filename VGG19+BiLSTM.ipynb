# Image Preprocessing

````python
from tensorflow.keras.preprocessing.image import load_img,img_to_array
from tensorflow.keras.applications.vgg19 import preprocess_input
import os
````

````python
final['Image_path']
````

````python
count=0
img_input_sequences=[]
def image_preprocessing(image_path):
    global count
    # Load the image
    img=load_img(image_path,target_size=(224,224))
    # convert pixels to numpy array
    img=img_to_array(img)
    # reshape the image for the model
    img=img.reshape((img.shape[0],img.shape[1],img.shape[2]))
    # preprocess the image
    img=preprocess_input(img)
    count=count+1
    print(count)
    img_input_sequences.append(img)
````

````python
final['Image_path'].apply(image_preprocessing)
````

# Merge two model

````python
from tensorflow.keras.layers import add
from tensorflow.keras.models import Model
````

````python
# image layers
ft1 = Flatten()(conv_base.output)
ft2 = Dense(256,activation='relu')(ft1)

# Question Layers
input1 = Input(shape=(max_len,))
sq1 = Embedding(vocab_size,256,input_length=max_len)(input1)
sq2 = LSTM(256)(sq1)

# Merge
mg1 = add([ft2, sq2])
mg2 = Dense(256, activation='relu')(mg1)
output = Dense(vocab_size, activation='softmax')(mg2)

model_vqa = Model(inputs=[conv_base.input, input1], outputs=output)
model_vqa.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model_vqa.summary()
````

````python
model_vqa.fit([img_input_sequences_array,X],y,epochs=5)
````

````python
def image(image_path):
    img=load_img(image_path,target_size=(224,224))
    img=img_to_array(img)
    img=img.reshape((img.shape[0],img.shape[1],img.shape[2]))
    img=preprocess_input(img)
    #print(img)
    #print(img.shape)
    #print(type(img))
    return img
def question(text):
    input_sequences = tokenizer.texts_to_sequences([text])
    #print(input_sequences)
    padded_input_sequences = pad_sequences(input_sequences,maxlen = max_len, padding = 'pre' )
    #print(padded_input_sequences)
    return padded_input_sequences
    
````

````python
# Assuming you have the image and question data
imag = image('synpic100173.jpg')
ques = question('মাথা কেটে ফেলতে হবে কিনা বুঝতে পারছি না, মাথায় কি হলো')

# Reshape the data to have a batch dimension
imag = np.expand_dims(imag, axis=0)  # Add a batch dimension
ques = np.array(ques)  # Ensure it's a numpy array

# predict using model_vqa instead of the sequential model
predictions = model_vqa.predict([imag, ques])
predictions
print(type(predictions))
# Get the index of the highest probability
#highest_probability_index = np.argmax(predictions)
pos = np.argmax(predictions)
print(pos)
for word,index in tokenizer.word_index.items():
        if index == pos:
            print(word)
            break
````
